```python
import json
import pandas as pd
```

```python
with open("depo4.json", "r") as f:
    json_data = json.load(f)

# ambil khusus key "data"
df = pd.DataFrame(json_data["data_container"])
cols_check = ["cargo", "container_size", "grade", "no_container", "state", "tgl_status", "tier", "updated_at", "vvoy"]
mask = df[cols_check].apply(lambda row: all(val == "-" or val is None for val in row), axis=1)
filtered_df = df[~mask].copy()
filtered_df["container_size"] = filtered_df["container_size"].replace("-", "20 DC")

for col in ["column"]:
    filtered_df[col] = pd.to_numeric(filtered_df[col], errors="coerce")

df_40hc = filtered_df[filtered_df["container_size"] == "40 HC"].copy()
df_non_40hc = filtered_df[filtered_df["container_size"] != "40 HC"].copy()

df_40hc = (
    df_40hc.sort_values("column")
    .groupby("no_container", as_index=False)
    .first()
)

filtered_df = pd.concat([df_non_40hc, df_40hc], ignore_index=True)        
filtered_df["column"] = filtered_df["column"].astype(str)
print(filtered_df.dtypes)
print("dapat data", len(filtered_df))

import random
import string

# contoh fungsi untuk ubah no_container
def randomize_no_container(value):
    if value == "-":
        return value
    
    # generate 4 huruf random awal & akhir
    prefix = ''.join(random.choices(string.ascii_uppercase, k=4))
    suffix = ''.join(random.choices(string.ascii_uppercase, k=4))
    
    # ambil bagian tengah (tanpa 4 depan & 4 belakang)
    if len(value) > 8:
        middle = value[4:-4]
    else:
        middle = ""
    
    return prefix + middle + suffix


df["no_container_new"] = df["no_container"].apply(randomize_no_container)
```
